```python
import matplotlib.pyplot as plt
xAxis=list(range(1,11))
yAxis=[12,23,34,45,56,43,54,55,75,86]
plt.plot(xAxis,yAxis,marker="o",linestyle="dashed",markerfacecolor="black",alpha=0.5)
plt.xlabel("numbers")
plt.ylabel("Range")
plt.show()
```

```python
flowers=["rose","jasmine","champa","lotus"]
counts=[23,32,43,43]
plt.bar(flowers,counts,color="red",edgecolor="black")
plt.show()
```

```python
flowers=["rose","jasmine","champa","lotus"]
counts=[23,32,43,43]
myColors=["red","grey","yellow","pink"]
myexplode=[0.2,0.5,0,0]
plt.pie(counts,labels=flowers,colors=myColors,autopct='%.2f',explode=myexplode,shadow=True)
plt.legend(title="flowers:")
plt.show()
```
